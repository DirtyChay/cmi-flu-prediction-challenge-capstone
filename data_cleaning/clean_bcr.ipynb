```python
# Import required libraries
import os
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv('../data/train_bcr.tsv', sep='\t')
columns_to_drop = ['event_id', 'sequence_id', 'sequence', 'rev_comp', 'productive', 
                   'vj_in_frame', 'stop_codon', 'v_germline_start', 'seqorient', 
                   'species', 'pcr_target_locus', 'single_cell', 'collapse_pass']
df = df.drop(columns=columns_to_drop)

# Clean biological_sex column (Standardize Male/Female)
df['biological_sex'] = df['biological_sex'].str.capitalize()

df.head()
```

```python
# Print all columns and their unique values
for col in df.columns:
    print(f"{col}: {df[col].unique()}\n")
```

```python
# Save the dataset
os.makedirs('../cleaned_data', exist_ok=True)
df.to_csv('../cleaned_data/bcr_cleaned.csv', index=False)
```
