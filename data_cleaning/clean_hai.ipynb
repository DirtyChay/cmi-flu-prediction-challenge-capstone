```python
# Import required libraries
import os
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv('../data/train_hai.tsv', sep='\t')
df = df.drop(columns=['hai_id'])
df.head()
```

```python
df['value'] = pd.to_numeric(df['value'], errors='coerce')
df['material'] = df['material'].replace({'specimen type: unknown': 'Unknown'})
```

```python
# Print all columns and their unique values
for col in df.columns:
    print(f"{col}: {df[col].unique()}\n")
```

```python
# Save the cleaned dataset to the cleaned_data folder
os.makedirs('../cleaned_data', exist_ok=True)
df.to_csv('../cleaned_data/hai_cleaned.csv', index=False)
```
