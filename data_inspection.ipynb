# Data Inspection

This notebook loads the raw tables and inspects them for quality issues.
Findings here inform the cleaning steps applied in the companion
`data_cleaning.ipynb` notebook.

```python
import pandas as pd
```

## 1. Load Raw Data

```python
INPUT_DATA_PATH = 'data/input/PART2-26-02-13_reorg'

train_hai = pd.read_csv(INPUT_DATA_PATH + '/train_hai.tsv', sep='\t')
train_participants = pd.read_csv(INPUT_DATA_PATH + '/train_participants.tsv', sep='\t')
```

```python
name_df = {
    'train_hai': train_hai,
    'train_participants': train_participants,
}
```

## 2. Inspect Raw Data

```python
for name, df in name_df.items():
    print(f"\n{'=' * 60}")
    print(f'TABLE: {name}  —  {df.shape[0]} rows, {df.shape[1]} columns')
    print(f"{'=' * 60}")
    print(df.dtypes, '\n')
    print('Missing values (%):')
    print((df.isna().mean() * 100).round(2))
    display(df.head(5))
```

### Value counts for categorical columns
Helps catch inconsistent labels, typos, or placeholder values.

```python
# Treat object and category dtypes as categorical
for name, df in name_df.items():
    cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
    if not cat_cols:
        print(f'{name}: no categorical columns found\n')
        continue
    for col in cat_cols:
        print(f"\n— {name}.{col} ({df[col].nunique()} unique) —")
        print(df[col].value_counts(dropna=False).head(15))
```

### Descriptive stats for numeric columns
Check for suspicious ranges (negative ages, extreme titers, etc.).

```python
for name, df in name_df.items():
    num_cols = df.select_dtypes(include='number').columns.tolist()
    if not num_cols:
        print(f'{name}: no numeric columns found\n')
        continue
    print(f"\n— {name} —")
    display(df[num_cols].describe())
```

### Unique counts
How many distinct participants, strains, timepoints, etc. are in each table?

```python
for name, df in name_df.items():
    print(f"\n— {name} —")
    for col in df.columns:
        print(f'  {col}: {df[col].nunique()} unique')
```

### Duplicate check
Look for fully duplicated rows in each table.

```python
for name, df in name_df.items():
    n_dupes = df.duplicated().sum()
    print(f'{name}: {n_dupes} duplicate rows'
          f' ({n_dupes / len(df) * 100:.2f}%)')
```

### Participant ID overlap between tables
Check whether all participants in one table appear in the other.

```python
# Collect participant_id sets from any table that has the column
pid_sets = {}
for name, df in name_df.items():
    if 'participant_id' in df.columns:
        pid_sets[name] = set(df['participant_id'].dropna().unique())

names = list(pid_sets.keys())
for i, a in enumerate(names):
    for b in names[i + 1:]:
        only_a = pid_sets[a] - pid_sets[b]
        only_b = pid_sets[b] - pid_sets[a]
        shared = pid_sets[a] & pid_sets[b]
        print(f'{a} vs {b}:')
        print(f'  shared:        {len(shared)}')
        print(f'  only in {a}: {len(only_a)}')
        print(f'  only in {b}: {len(only_b)}')
```
